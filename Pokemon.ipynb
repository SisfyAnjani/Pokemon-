**Arti kolom :**  [sumber](https://m.kaskus.co.id/thread/5efb8f07dbf7643b483fb9c0/pokemon-battle-basics-part-2-pokemon-stats-and-nature)

1. **'#'** -> Kode Jenis Pokemon

2. **Name** -> Nama Pokemon

3. **Type 1** -> Tipe pertama pokemon

4. **Type 2** -> Tipe Kedua pokemon

5. **Total** -> HP + Attack + Defense + Sp.Atk + Sp.Def + Speed

6. **HP (Hit Points)** -> jumlah kekuatan yang dimiliki pokemon (jumlah ini menentukan seberapa lama pokemon bisa bertahan dalam pertarungan)

7. **Attack** -> Serangan yang dilakukan oleh Pokemon untuk menyerang lawannya.

8. **Defense** -> Pertahanan Pokemon ketika di serang.

9. **Sp. Atk (Special Attack)** -> Pengaruh Kekuatan Special damage dari Pokemon ketika menyerang.

10. **Sp. Def (Special Defense)** -> Pengaruh pertahanan terhadap Special damage dari pokemon ketika diserang.

11. **Speed** -> mempengaruhi kecepatan pokemon (menentukan pokemon mana yang dapat giliran pertama menyerang)

12. **Generation** -> membedakan yang baru dan lama sesuai masa rilis.

13. **Legendary** -> Pokemon yang langka di dunia pokemon.



> ada 386 data NAN di kolom Type 2





*   Pokemon memiliki 2 tipe pada data diatas, tetapi pada tipe 2 terdapat 386 data yang NAN artinya, ada 386 Pokemon yang hanya memiliki 1 Tipe. maka akan hipotesiskan bahwa kolom Type 2 yang kosong akan diisi dengan nilai 0 yang artinya tidak memiliki tipe ke2.




```python
# Mengisi nilai NaN dalam kolom 'kolom Type 2' dengan 0
pokemon['Type 2'] = pokemon['Type 2'].fillna(0)

# Mengisi nilai NaN dalam seluruh DataFrame dengan 0
df = pokemon.fillna(0)
df
```

## Feature Engineering + Visualisasi

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Menghitung matriks korelasi
korelasi = numeric_columns.corr()

# Membuat heatmap dari matriks korelasi
sns.heatmap(korelasi, annot=True, cmap='Blues')
plt.show()
```

### Pokemon yang langka

```python
pokemon_langkah = pokemon[pokemon['Legendary']==True]
pokemon_langkah
```

```python
# Membuat scatter plot dengan seaborn
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Attack', y='Defense', data=pokemon_langkah, hue='Name', palette='viridis', s=100)
plt.xlabel('Attack')
plt.ylabel('Defense')
plt.title('Attack vs Defense of Legendary PokÃ©mon')
plt.legend(loc='upper left')
plt.show()
```

### total tertinggi

```python
pokemon_maxtotal = pokemon.sort_values(by='Total', ascending=False)
pokemon_maxtotal
```

```python
#visualisasi 10 pokemon tertinggi
pokemon_maxtotal = pokemon_maxtotal.head(10)
pokemon_maxtotal

# Membuat bar plot dengan matplotlib
plt.figure(figsize=(12, 8))
plt.bar(pokemon_maxtotal['Name'], pokemon_maxtotal['Total'], color='skyblue')
plt.xlabel('Name')
plt.ylabel('Total Stats')
plt.title('PokÃ©mon dengan Total Stats Tertinggi')
plt.xticks(rotation=45)
plt.show()
```
